```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
```

```python
print(dir(digits))
```

```python
print(type(digits.data))
```

```python
import pandas as pd
```

```python
#  Create a figure with 1 row and 5 columns
fig, axes = plt.subplots(1, 5, figsize=(5, 2))

for i in range(5):
    # 2. Use the specific 'axis' for each image
    axes[i].matshow(digits.images[i], cmap='gray')
    
    # 3. Optional: Remove ticks for a cleaner look
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].set_title(f"Label: {digits.target[i]}")

plt.tight_layout() # Prevents titles from overlapping
plt.show()
```

```python
df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
df.shape
```

```python
input = df.drop('target', axis='columns')
```

```python
input.head()
```

```python
label = df['target']
label.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, label, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=10.00,
    kernel='rbf',
    degree=3,
    gamma='scale')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
